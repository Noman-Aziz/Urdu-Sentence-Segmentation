**ASSIGNMENT REPORT**



1.   How did i detect patterns from text?
     *   In urdu language, the sentence mostly end with combination of verbs and new line words.
     *   If new line word comes at the end and after it an conjunction is used, then it means that the sentence is continued and not ended.

2.   Can i identify end-words of a sentence in urdu?
     *   I tried to maintain a list of new line words, conjunction words, verb words in my code.
     *   I try to find a new-line word by first converting the passsage into tokens after normalizing it with urduhack "normalize" function to fix any problems like 'ﺁ' and 'ﺂ' are to be meant as a single character and are replaced by 'آ'.
     *   Furthermore, once a new-line word is matched, i check that the next word in the tokens is not a conjunction word, apart from this i also check that the next word is also not a newline word, also i loosely check that the previous word is a verb word (it makes sure to a point that the current word markes the end of a sentence).
     *   If above conditions are met, i combine the current and previous words to make them a sentence and store it in a container.

3.   How can i evaluate my approach?
     *   Self Based : from what i can tell, this method needs a full authentic list of urdu verbs as well as newline and conjunction words, whereas my code only contains few of these words. Therefore, my code will segment the sentences correct at some places and will not at most places. Moreover, i think my code is more efficient than the urduhack sentence segmentation code.
     *   Code Based : 
         *   I will first try to count the number of dashes in the self segmented passage and the number of dashes in the passage that my code has segmented. This will give me a percentage P1 of number of sentences that are segmented.
         *   Then, i will compare the end words where dash is applied with the original location of dashes and generate a percentage P2.
         *   Then, i will compare the individual sentences segmented with the actual segmented sentences to caclulate individual percentage of length matched and take avg of all in the end.
             * i.e, *P3(n) = ( ( correct length matched of n sentence) / (actual length of n sentence) ) * 100* 
             * i.e, *P3 = ( P3(1) + ... + P3(n) ) / n* 
         *   Then, finally i will calculate the avg of all three percentages to evaluate my code by,
             * i.e, *P = ( P1 + P2 + P3 ) / 3* 





```python
from urduhack.normalization import normalize

#Checks if word is a verb word
def isVerbWord(word, verbWords):

  for verb in verbWords:
    if verb in word:
      return True

  return False

#Checks if a word is a newline word
def isEndingWord(word, index, tokens, newlineWords, conjunctionWords, verbWords):

    for newlineWord in newlineWords:

      if (word.endswith(newlineWord)) and (index + 1 < len(tokens)) and (tokens[index+1] not in conjunctionWords): 

        #checks if previous word is a verb word
        if isVerbWord(tokens[index-1], verbWords):
          return True

        #checks if next word is not a newline word
        elif not isEndingWord(tokens[index+1], index + 1, tokens, newlineWords, conjunctionWords, verbWords):
          #checks if comma does not exist in next word
          if not isEndingWord("،", index + 1, tokens, newlineWords, conjunctionWords, verbWords):
            return True

    return False

def segmentSentence(passage):

  #SOME USEFUL VARIABLES
  conjunctionWords = ['جنہیں', 'جس', 'جن', 'جو', 'اور', 'اگر', 'اگرچہ', 'لیکن', 'مگر', 'پر', 'یا', 'تاہم', 'کہ', 'کر',
                        'تو', 'گے', 'گی','پہلے','کیونکہ','چونکہ','ورنہ','بلکہ','تاکہ', 'جیسے','چنانچہ']

  #Words that denote end of a sentence
  newlineWords = ['کیجیے', 'کیجئے', 'گئیں', 'تھیں', 'ہوں', 'خریدا', 'گے', 'ہونگے', 'گا', 'چاہیے', 'ہوئیں', 'گی',
                        'تھا', 'تھی', 'تھے', 'ہیں', 'ہے','سکے']

  verbWords = [ 'کرنا', 'دوڑانہ', 'کھیلنا', 'گھورنا', 'دینا', 'چڑھانا', 'دیکھنا', 'چمکنا', 'جلنا', 'چلنا', 'گھمانا', 'رگڑنا', 'سننا' ]

  #normalizing the passage
  normalizedPassage = normalize(passage)

  print(normalizedPassage)

  #tokenizing the normalized passage into words
  tokens = normalizedPassage.split(' ')

  #print(tokens)

  #empty container to hold sentences
  sentences = []
  
  #segmenting the sentences
  sentenceFound = False
  sentence = ""

  for index, word in enumerate(tokens):

    #Check for end of sentence words
    if isEndingWord(word, index, tokens, newlineWords, conjunctionWords, verbWords):
        sentenceFound = True

    #Add the current word in sentence container
    if sentence == "":
      sentence += word
    else:
      sentence = sentence + " " + word
      if sentenceFound:
        sentence += "۔"

    #Reset the sentence container and add it into sentences container
    if (sentenceFound is True) or ('۔' in word) or ('؟' in word):
      sentenceFound = False
      sentences.append(sentence)
      sentence = ""

  return sentences

#  for sentence in sentences:
#    print(sentence)
```
